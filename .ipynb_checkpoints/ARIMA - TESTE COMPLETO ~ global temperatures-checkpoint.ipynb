```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
import statsmodels
import statsmodels.api as sm
import seaborn as sns
from sklearn.metrics import mean_squared_error
from statsmodels.stats.stattools import jarque_bera
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
import datetime
%matplotlib inline
from pmdarima.arima import auto_arima
```

```python
# Carregando os dados:
df=pd.read_csv('Sunspots.csv')
# reduzindo amostra:
df_sunspots = df[2065:3265]
df_sunspots = df_sunspots.set_index('Date')
df_sunspots.index = pd.to_datetime(df_sunspots.index, format='%Y-%m-%d')
df_sunspots = df_sunspots[['Monthly Mean Total Sunspot Number']].groupby('Date').sum()
df_sunspots.rename(columns={'Monthly Mean Total Sunspot Number': 'sunspots'}, inplace = True)

df_sunspots.head(2)
```

```python
# decompondo a srie:
seasonplot = seasonal_decompose(df_sunspots)
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(15,8))
seasonplot.observed.plot(ax=ax1, ylabel='Observaes') 
seasonplot.trend.plot(ax=ax2, ylabel='Tendncia') 
seasonplot.seasonal.plot(ax=ax3, ylabel='Sazonalidade')
seasonplot.resid.plot(ax=ax4, ylabel='Rudo') 
plt.tight_layout()
```

```python
# aplicando o teste Dickey-Fuller Aumentado (ADF) para verificar estacionariedade na srie:

# Extraindo a coluna que contm os dados da srie temporal
dados_serie = df_sunspots['sunspots']
resultado_adf = adfuller(dados_serie)

# Extraindo os resultados do teste ADF
valor_p = resultado_adf[1]
estatistica_teste = resultado_adf[0]

# Comparando o valor-p com um nvel de significncia (geralmente 0.05)
nivel_de_significancia = 0.05

if valor_p <= nivel_de_significancia:
    print("A srie temporal  estacionria (rejeita-se a hiptese nula).")
else:
    print("A srie temporal no  estacionria (no se rejeita a hiptese nula).")

# Exiba a estatstica do teste
print(f"Estatstica do teste ADF: {estatistica_teste}")
print(f"Valor-p: {valor_p}")
```

```python
# verificando autocorrelao e autocorrelao parcial:
# auto-regresso (p) = autorrelao; mdia mvel (q)= autocorrelao parcial.
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15,10))
sm.graphics.tsa.plot_acf(df_sunspots.values.squeeze(), lags=40, ax=ax1)
sm.graphics.tsa.plot_pacf(df_sunspots.values.squeeze(), lags= 40, ax=ax2, method='ywm')
plt.show()
```

```python
# Aplicando mtodo Auto ARIMA (j pode-se definir max e min valores dos parmetros para facilitar o cdigo):
# A sazonalidade, se mensal, usar m=12; Por padro o P iniciar em 0 e o mximo ser 2; o Q por padro ser 1 e o mximo, por padro, ser 2; o d ser 1 e o mximo, por padro, ser 2; o D ser 1, por padro j  1.
# Sero gerados vrios modelos ARIMA e os seus respectivos AIC's e BICS (critrios de informao).
# Stepwise permite diminuir o nmero de modelos gerados, deixa o cdigo mais leve.
stepwise=auto_arima(df_sunspots,start_p=0, start_q=0,max_p=3,max_q=2,m=12,start_P=0,
                     seasonal=True,d=0,D=1,trace=True, error_action='ignore',suppress_warnings=True, stepwise=True)
```

```python
# funo para calcular o erro dos modelos
from math import sqrt
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_error

def check_erro(orig, prev, nome_col='', nome_indice=''):
    
    vies = np.mean(orig - prev)
    mse = mean_squared_error(orig, prev)
    rmse = sqrt(mean_squared_error(orig, prev))
    mae = mean_absolute_error(orig, prev)
    mape = np.mean(np.abs((orig - prev) / orig)) * 100
    
    grupo_erro = [vies, mse, rmse, mae, mape]
    serie = pd.DataFrame(grupo_erro, index=['VIS','MSE','RMSE','MAE', 'MAPE'], columns=[nome_col])
    serie.index.name = nome_indice
    
    return serie
```

```python
def plotar_erro(dados, figsize=(18,8)):
    
    # Criando a coluna do erro
    dados['Erro'] = dados.iloc[:,0] - dados.iloc[:,1]
    
    plt.figure(figsize=figsize)
    ax1 = plt.subplot2grid((2,2), (0,0))
    ax2 = plt.subplot2grid((2,2), (0,1))
    ax3 = plt.subplot2grid((2,2), (1,0))
    ax4 = plt.subplot2grid((2,2), (1,1))
    
    #Plotando os valores reais e previstos
    ax1.plot(dados.iloc[:,0:2])
    ax1.legend(['Real','Prev'])
    ax1.set_title('Valores Reais vs Previstos')
    
    # Erro vs Valores previstos
    ax2.scatter(dados.iloc[:,1], dados.iloc[:,2])
    ax2.set_xlabel('Valores Previstos')
    ax2.set_ylabel('Resduo')
    ax2.set_title('Resduo vs Valores Previstos')
    
    ## QQ Plot do erro
    sm.graphics.qqplot(dados.iloc[:,2], line='r', ax=ax3)
    
    # Grfico de autocorrelao do Erro
    plot_acf(dados.iloc[:,2], lags=60,zero=False, ax=ax4)
    plt.tight_layout()
    plt.show()
```
