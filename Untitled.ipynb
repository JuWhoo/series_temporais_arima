```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
import statsmodels
import statsmodels.api as sm
import seaborn as sns
from sklearn.metrics import mean_squared_error
from statsmodels.stats.stattools import jarque_bera
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
import datetime
%matplotlib inline
from pmdarima.arima import auto_arima
```

```python
# Carregando os dados:
df=pd.read_csv('global_temp_monthly.csv')

# Carregando os dados:

df_temp=df.dropna()
df_temp= df_temp.set_index(['Date'])

df_temp.index = pd.to_datetime(df_temp.index, format='%Y-%m')
df_temp = df_temp[['Mean']].groupby('Date').sum()
df_temp.rename(columns={'Mean': 'Temperature'}, inplace = True)

df_temp.head(10)
```

```python
df_temp.iloc[[0, -1]]
```

```python
# decompondo a srie

seasonplot = seasonal_decompose(df_temp)
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(15,8))
seasonplot.observed.plot(ax=ax1, ylabel='Observaes') 
seasonplot.trend.plot(ax=ax2, ylabel='Tendncia') 
seasonplot.seasonal.plot(ax=ax3, ylabel='Sazonalidade')
seasonplot.resid.plot(ax=ax4, ylabel='Rudo') 
plt.tight_layout()
```

```python
# aplicando o teste Dickey-Fuller Aumentado (ADF) para verificar estacionariedade na srie:

# Extraindo a coluna que contm os dados da srie temporal
dados_serie = df_temp['Temperature']
resultado_adf = adfuller(dados_serie)

# Extraindo os resultados do teste ADF
valor_p = resultado_adf[1]
estatistica_teste = resultado_adf[0]

# Comparando o valor-p com um nvel de significncia (geralmente 0.05)
nivel_de_significancia = 0.05

if valor_p <= nivel_de_significancia:
    print("A srie temporal  estacionria (rejeita-se a hiptese nula).")
else:
    print("A srie temporal no  estacionria (no se rejeita a hiptese nula).")

# Exiba a estatstica do teste
print(f"Estatstica do teste ADF: {estatistica_teste}")
print(f"Valor-p: {valor_p}")
```

```python
# em caso negativo, para teste do melhor resultado da srie para se tornar estacionria, pode ser aplicado a diferenciao:
diff_1 = df_temp.diff().dropna()
# testar novamente por ADF
dados_serie = diff_1['Temperature']
resultado_adf = adfuller(dados_serie)

# Extraindo os resultados do teste ADF
valor_p = resultado_adf[1]
estatistica_teste = resultado_adf[0]

# Comparando o valor-p com um nvel de significncia (geralmente 0.05)
nivel_de_significancia = 0.05

if valor_p <= nivel_de_significancia:
    print("A srie temporal  estacionria (rejeita-se a hiptese nula).")
else:
    print("A srie temporal no  estacionria (no se rejeita a hiptese nula).")

# Exiba a estatstica do teste
print(f"Estatstica do teste ADF: {estatistica_teste}")
print(f"Valor-p: {valor_p}")
```
